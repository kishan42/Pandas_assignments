Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

windows function is used in pandas to perform or calculate the statistic value like mean , max , min , mod , sum across the set of rows that are interrelated with current row.

Windows function in Pandas can be broadly divided into three categories, namely- Aggregate, Ranking, and Value.

The Aggregate category of window functions can be of three types, namely-

Group,
Rolling, and
Expanding.
The Ranking category of window functions can be of five types, namely-

Row Numbers,
Row numbers if of two types:
reset_index()
cumcount()
Rank,
Rank is of four types:
default_rank
min_rank
NA_bottom
descending
Dense rank,
Percent, and
N-Tile / qcut().
The Value category of window functions can be of two types, namely-

Lag / Lead, and
First / Last/nth.

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
a = input('Enter the file path :')
b = input('Enter the column name :')
c = list(input('Enter the category order :').split())
print(a,b,c)
```

```python
df3 = pd.read_csv(a)
```

```python
df3
```

```python
df3[b] = pd.Categorical(df3[b])
```

```python
df3.info()
```

```python
df3[b] = df3[b].cat.reorder_categories(c)
```

```python
df3.sort_values(by=[b])
```
